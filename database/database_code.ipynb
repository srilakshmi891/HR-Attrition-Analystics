```python
conn=sqlite3.connect("hr_attrition.db")
cursor=conn.cursor()
print("Database Connected Successfully")
```

```python
query = """
SELECT COUNT(*) AS total_employees
FROM employees
"""
```

```python
result = pd.read_sql(query, conn)

print(result)
```

```python
query2 = """
SELECT COUNT(*) AS attrition_count
FROM employees
WHERE Attrition = 1
"""
```

```python
result2 = pd.read_sql(query2, conn)

print("\nAttrition Employees")
print(result2)
```

```python
query3 = """
SELECT 
ROUND(
    (SUM(Attrition) * 100.0) / COUNT(*), 2
) AS attrition_rate
FROM employees
"""
```

```python
result3 = pd.read_sql(query3, conn)

print("\nAttrition Rate")
print(result3)
```

```python
query4 = """
SELECT 
Department,
COUNT(*) AS total_employees,
SUM(Attrition) AS attrition_employees
FROM employees
GROUP BY Department
"""
```

```python
result4 = pd.read_sql(query4, conn)

print("\n JobSatisfaction Wise Attrition")
print(result4)
```

```python

# ---------------------------------
# Average Salary by Department
# ---------------------------------
query5 = """
SELECT 
Department,
AVG(MonthlyIncome) AS avg_salary
FROM employees
GROUP BY Department 
"""
```

```python
result5 = pd.read_sql(query5, conn)

print("\nAverage Salary by Department")
print(result5)
```

```python
# ---------------------------------
# Overtime Analysis
# ---------------------------------

query6 = """
SELECT 
OverTime,
COUNT(*) AS total_employees,
SUM(Attrition) AS attrition_count
FROM employees
GROUP BY OverTime
"""
```

```python
result6 = pd.read_sql(query6, conn)

print("\nOvertime Analysis")
print(result6)
```

```python
# ---------------------------------
# Job Satisfaction Analysis
# ---------------------------------
query7 = """
SELECT 
JobSatisfaction,
COUNT(*) AS employees
FROM employees
GROUP BY JobSatisfaction
ORDER BY JobSatisfaction
"""
```

```python
result7 = pd.read_sql(query7, conn)

print("\nJob Satisfaction Analysis")
print(result7)
```

```python
# ---------------------------------
# Gender Wise Attrition
# ---------------------------------

query8 = """
SELECT 
Gender,
COUNT(*) AS total_employees,
SUM(Attrition) AS attrition_count
FROM employees
GROUP BY Gender
"""
```

```python
result8 = pd.read_sql(query8, conn)

print("\nGender Wise Attrition")
print(result8)
```

```python
# ---------------------------------
# Top 10 Highest Salary Employees
# ---------------------------------

query9 = """
SELECT 
MonthlyIncome,
JobLevel,
TotalWorkingYears
FROM employees
ORDER BY MonthlyIncome DESC
LIMIT 10
"""
```

```python
result9 = pd.read_sql(query9, conn)

print("\nTop 10 Highest Salary Employees")
print(result9)
```

```python
# ---------------------------------
# Average Experience
# ---------------------------------

query10 = """
SELECT 
AVG(TotalWorkingYears) AS avg_experience
FROM employees
"""
```

```python
result10 = pd.read_sql(query10, conn)

print("\nAverage Experience")
print(result10)
```
